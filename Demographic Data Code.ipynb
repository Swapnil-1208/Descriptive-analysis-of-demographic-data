```python
df2020=df[df['Year']==2020]  #Fetching only records where Year=2020 for the first three tasks
df2020.drop(['Year'],axis=1,inplace=True)  #dropping the year columns for the first three task
```

# TASK-1 

# FREQUENCY DISTRIBUTION OF THE VARIABLES

FOR INDEPENDENT NUMERICAL VARIABLE : Using Matplotlib and Seaborn Library(for better and detailed graph)

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 8))

fig.suptitle('Frequency Distribution of Numerical Variables')

sns.histplot(df2020['Total.Fertility.Rate'],stat='density', ax=axes[0, 0])
sns.histplot(df2020['Life.Expectancy.at.Birth..Females'],stat='density', ax=axes[0, 1])
sns.histplot(df2020['Life.Expectancy.at.Birth..Males'],stat='density', ax=axes[1, 0])
sns.histplot(df2020['Life.Expectancy.at.Birth..Both.Sexes'],stat='density', ax=axes[1, 1])

```

```python
a=plt.scatter(df2020['Life.Expectancy.at.Birth..Both.Sexes'], df2020['Life.Expectancy.at.Birth..Females'], color='r')
b=plt.scatter(df2020['Life.Expectancy.at.Birth..Both.Sexes'], df2020['Life.Expectancy.at.Birth..Males'], color='b')
plt.xlabel('Life Expectancy Range')
plt.ylabel('Life Expectancy')
plt.title('Differences between the sexes')
plt.legend((a,b),('Females','Males'))
plt.show()
```

# TASK-3

# Comparison of variability within and among sub-region

```python
sns.set(rc={'figure.figsize':(8,8)})
```

```python
my_pal = sns.color_palette("bright")
```

```python
ax = sns.boxplot(data=(df2020.sort_values(["Region"])),x='Total.Fertility.Rate', y='Subregion', hue='Region', palette=my_pal)
ax.set(xlabel='Total Fertility Rate', ylabel='Sub Regions')
```

```python
ax = sns.boxplot(data=(df2020.sort_values(["Region"])),x='Life.Expectancy.at.Birth..Both.Sexes', y='Subregion', hue='Region', palette=my_pal)
ax.set(xlabel='Life Expectancy at Birth : Both Sexes', ylabel='Sub Regions')
```

```python
ax = sns.boxplot(data=(df2020.sort_values(["Region"])),x='Life.Expectancy.at.Birth..Males', y='Subregion', hue='Region', palette=my_pal)
ax.set(xlabel='Life Expectancy at Birth : Males', ylabel='Sub Regions')
```

```python
ax = sns.boxplot(data=(df2020.sort_values(["Region"])),x='Life.Expectancy.at.Birth..Females', y='Subregion', hue='Region', palette=my_pal)
ax.set(xlabel='Life Expectancy at Birth : Females', ylabel='Sub Regions')
```

# YEAR COMPARISON

```python
year_comparison=df.pivot_table(aggfunc='mean', values=['Total.Fertility.Rate','Life.Expectancy.at.Birth..Both.Sexes','Life.Expectancy.at.Birth..Females','Life.Expectancy.at.Birth..Males'],dropna=True,index=['Year'])
year_comparison
```
